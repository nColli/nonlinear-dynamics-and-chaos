### Función de Lyapunov para analizar comportamiento de Lokta-Volterra

$$V(x,y) = (x - 3 \ln(x)) + (y - 2 \ln(y))$$

Graficar V(x,y)=k, con k>0, y verificar que son todas curvas cerradas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Función de Lyapunov para que se pueda evaluar:

```python
def v(x,y):
    return x-3*np.log(x)+y-2*np.log(y)
```

**Puntos en los que se va a evaluar la función:**<br>
creo un meshgrid para ordenar los puntos en una grilla y poder evaluarlos

```python
x = np.arange(0.1, 20, 0.25)
y = np.arange(0.1, 20, 0.25)
X, Y = np.meshgrid(x, y) #crea una grilla de puntos en base a x,y para evaluar la función v en cada punto
```

Evaluo la función en esa grilla X,Y

```python
z = v(X, Y)
```

z es una matriz donde esta los resultados de la funcion v evaluada en cada punto x,y.<br>
Osea Z[0,0] = V(X[0,0], Y[0,0]) <br>
Asi con cada punto.

**Punto de equilibrio**<br>
Defino al punto de equilibrio que satisface sistema sea igual a 0.<br>
Calculo se hace despejando sist. de ecuación.<br>
Evaluo funcion v en ese punto para tomarlo como valor inicial para las curvas de nivel.

```python
pe_x, pe_y = 3, 2
pe_v = v(pe_x, pe_y)
```

**Curvas de nivel**
Defino las curvas de nivel, osea los k con los que igualo V(x,y).<br>
V(x,y)=K, con k>0

```python
k = np.arange(0.5, 10, 0.5)
```

**Grafico**
Punto de equilibrio para analizar como las curvas de nivel orbitan el PE.<br>
Curvas de nivel con plt.contour que permite graficarlas directamente.

```python
plt.plot(pe_x, pe_y, 'ro') #punto de equilibrio

cn = plt.contour(X, Y, z, levels=k) #curvas de nivel de v
plt.clabel(cn, fontsize=8)
```
