### Operaciones con números complejos y matrices

```python
# numeros complejos - j en vez de i
z1 = 3 + 2j
z2 = 1 - 4j

z1 + z2  # suma de complejos
```

```python
import numpy as np

np.exp(z1)
```

```python
abs(z1)  # modulo de un numero complejo
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

A.dot(B)  # producto matricial - tmb np.dot(A, B)
```

```python
# slicing
a = np.array([1, 2, 3, 4, 5])

a[1:4]  # incluye a[1], excluye a[4]
```

```python
3 in a, 10 in a
```

### Algebra lineal con Numpy

```python
import numpy.linalg as la

la.eig(A) # A - matriz
```

```python
eigenvalues, eigenvectors = la.eig(A)
eigenvalues, eigenvectors
```

```python
la.inv(A)
```

### Matplotlib

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y = np.sin(x)

plt.plot(x, y)
```

```python
t = np.linspace(0, 100, 1000)
x1 = np.sin(t)*np.exp(-0.1*t)
x2 = np.cos(t)

plt.plot(t, x1, label='x1')
plt.plot(t, x2, label='x2', linewidth=1)
plt.legend()

plt.xlabel('t')
plt.ylabel('x1, x2')
plt.title('Comparación oscilaciones')
```
