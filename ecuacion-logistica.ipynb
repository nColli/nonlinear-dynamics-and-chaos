# Ecuación Logística

$P(t)$ es la población de peces a lo largo del tiempo t.<br>
$\beta$ es la tasa de nacimiento de la población.<br>
$\delta$ es la tasa de muerte de la población.<br>
$$\frac{dP}{dt} = P(t)(\beta - \delta P(t))$$

Ejemplo base parametros = 1<br>
Para entender la evolución de la función logística, grafico P y dp_dt.<br>
Con esto podría observar los PE de forma empírica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dp_dt(p, t):
    return p*(1-p)
```

```python
P = np.linspace(-0.5, 1.5, 1000)
dP = dp_dt(P, 0)
```

Flechas para diagrama de fases 1D

```python
P_flechas = np.linspace(-0.5, 1.5, 20)
dP_flechas = dp_dt(P_flechas, 0)
```

Diagrama de Fases

```python
plt.plot(P, dP, 'b-', linewidth=2, label='dP/dt = P(1 - P)')

# flechas
plt.quiver(
    P_flechas, np.zeros_like(P_flechas),
    dP_flechas, np.zeros_like(dP_flechas),
    scale=10, scale_units='inches',
    color='green', alpha=0.9, width=0.01
)

# Puntos de equilibrio
plt.plot(0, 0, 'ro', markersize=10, label='P = 0 (inestable)')
plt.plot(1, 0, 'gs', markersize=10, label='P = 1 (estable)')


plt.axhline(0, color='k', linewidth=0.8, linestyle='--')

plt.xlabel('P', fontsize=14)
plt.ylabel('dP/dt', fontsize=14)
plt.title('Ecuación logística: dP/dt = P(1 - P)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.ylim(-0.01, 0.3)
plt.show()
```
